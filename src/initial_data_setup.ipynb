```python
import pandas as pd
import requests 
import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
```

```python
original_df = pd.read_csv("https://projects.fivethirtyeight.com/nfl-api/nfl_elo.csv")
original_df
```

```python
original_df.info()
```

```python
original_df.columns
```

```python
# extract only columns that has impact 

X_features = [ 'team1', 'team2', 'elo1_pre', 'elo2_pre', 'elo_prob1', 'elo_prob2', 
       'qbelo1_pre', 'qbelo2_pre', 'qbelo_prob1', 'qbelo_prob2',
       'score1', 'score2']
```

```python
df = original_df[X_features].copy()
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.info()
```

```python
# derive the winning team
df['winner'] = np.where(df['score1'] > df['score2'], 1, 2)

df.to_csv("../data/final_nfl_elo.csv")
```
